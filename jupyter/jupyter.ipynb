```python
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
import numpy as np
```

```python
m = Basemap(projection='ortho',
            lat_0=38.5,
            lon_0=95,
            resolution='c',
            area_thresh=1000.0)
m.drawcoastlines()
m.drawcountries()
m.fillcontinents(color='#78D2F9')
x, y = m(104.065925, 30.657272,)
m.plot(x, y, 'ro', markersize=10)
plt.show()
```

```python
m = Basemap(
    projection='merc',
    lat_0=30.657272, 
    lon_0=104.065925,
    resolution='c',
    llcrnrlon=104.2178123382,
    llcrnrlat=30.4781772402,
    urcrnrlon=103.9213455517,
    urcrnrlat=30.7828453209,
    area_thresh=10
)
m.drawcoastlines()
m.drawcountries()
m.fillcontinents(color='#78D2F9')
m.drawmeridians(np.arange(103.9, 104.5, 0.1), color='#666666')
m.drawparallels(np.arange(30, 31, 0.1), color='#666666')
x, y = m(104.065925, 30.657272,)
m.plot(x, y, 'ro', markersize=10)
plt.show()
```
